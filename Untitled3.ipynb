```python
import cv2
import matplotlib.pyplot as plt

# 2) Carga de la imagen (ajusta el nombre de archivo)
img = cv2.imread('filter.jpeg')
if img is None:
    raise FileNotFoundError("No se encontr 'tu_imagen.jpg'. Verifica el nombre y la ruta.")

# 3) Convertir a escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 4) Umbral para aislar zonas muy blancas
#    Ajusta 'thr' segn tu imagen (0255). Aqu 240 solo deja pxeles casi blancos.
thr = 240
_, mask_white = cv2.threshold(gray, thr, 255, cv2.THRESH_BINARY)

# 5) Encontrar contornos de las zonas blancas
contours, _ = cv2.findContours(mask_white, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 6) Dibujar rectngulos alrededor del contorno ms grande (el ms blanco)
if contours:
    # Seleccionamos el contorno de mayor rea
    c = max(contours, key=cv2.contourArea)
    x,y,w,h = cv2.boundingRect(c)
    out = img.copy()
    cv2.rectangle(out, (x,y), (x+w, y+h), (0,255,0), 2)
else:
    out = img.copy()
    print("No se detectaron regiones suficientemente blancas.")

# 7) Mostrar resultados
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original (Color)')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gray, cmap='gray')
plt.title('Escala de grises')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
plt.title('Deteccin del ms blanco')
plt.axis('off')

plt.show()
```
